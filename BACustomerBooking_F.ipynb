```python
dff = df.rename(columns = {'Mobile':'sales_channel'})

dff
```

# Splitting the Data

```python
X = dff.drop('booking_complete', axis =  1)

y = dff['booking_complete']
```

# Training the Data

```python
from sklearn.preprocessing import LabelEncoder

la = LabelEncoder()

y =  la.fit_transform(y)


from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=1)
```

```python
print('X_train', X_train.shape)
print('X_test', X_test.shape)
print('y_train', y_train.shape)
print('y_test', y_test.shape)
```

# Building the Model


```python
from sklearn.ensemble import RandomForestClassifier

classifier = RandomForestClassifier(n_estimators = 4, criterion = 'entropy', random_state = 0)

classifier.fit(X_train, y_train)

y_pred = classifier.predict(X_test)
```

```python
dff.booking_complete.unique()
```

# Evaluating th Model

```python
from sklearn.metrics import confusion_matrix

from sklearn.metrics import classification_report, accuracy_score

BA = confusion_matrix(y_test, y_pred)

print(f'Confusion_matrix :\n \n{BA}') 
```

```python
print('Accuracy score :\n')
print(accuracy_score(y_test, y_pred))
```

```python
print('Classification report:\n')
print(classification_report(y_test, y_pred))
```
