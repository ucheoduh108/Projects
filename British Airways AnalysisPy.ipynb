```python
import requests

from bs4 import BeautifulSoup 

r = requests.get('https://www.airlinequality.com/airline-reviews/british-airways/')

c = r.content

soup = BeautifulSoup(c, 'html.parser')
#print(soup.prettify())

all = soup.find_all('div', {'class':'review-stats'})
```

```python
L = []

base_url = 'https://www.airlinequality.com/airline-reviews/british-airways/page/'  #to capture other pages
for page in range(0,372,1):   
    print(base_url + str(page))
    r = requests.get(base_url + str(page))
    c = r.content  #r.json()['list']
    soup = BeautifulSoup(c,'html.parser')
    #print(soup.prettify())
    all = soup.find_all('div', {'class':'review-stats'})
  

    for items in all:

            d = {}

            try:
                d['Aircraft']= items.find_all('td',{'class':'review-value'})[-6].text 
            except:
                d['Aircraft'] = None 

            try:
                d['Type_of_Traveller'] = items.find_all('td',{'class':'review-value'})[-5].text
            except:
                d['Type_of_Traveller'] = None
                
            try:
                d['Seat_type'] = items.find_all('td',{'class':'review-value'})[-4].text
            except:
                d['Seat_type'] = None

            try:
                d['Route'] = items.find_all('td',{'class':'review-value'})[-3].text
            except:
                d['Route'] = None

            try:
                d['Date_flown'] = items.find_all('td',{'class':'review-value'})[-2].text
            except:
                d['Date_flown'] = None

            try:
                d['Recommended'] = items.find_all('td',{'class':'review-value'})[-1].text
            except:
                d['Recommended'] = None
                
            try:
                TR = items.find_all('tr')
                for tr in TR:
                    tds = tr.find_all('td', {'class':'review-rating-header seat_comfort'})
                    if tds:
                        star_fill_values = [span.text for span in tr.find('td',{'class' :
                        'review-rating-stars stars'}).find_all('span',{'class':'star fill'})]
                        total_star_fill = len(star_fill_values)
                        d['Seat_comfort'] = total_star_fill
            except :
                d['Seat_comfort'] = None

            try:
                TR = items.find_all('tr')
                for tr in TR:
                    tds = tr.find_all('td', {'class':'review-rating-header cabin_staff_service'})
                    if tds:
                        star_fill_values = [span.text for span in tr.find('td',{'class' :
                        'review-rating-stars stars'}).find_all('span',{'class':'star fill'})]
                        total_star_fill = len(star_fill_values)
                        d['Cabin_service'] = total_star_fill
            except :
                d['Cabin_service'] = None
                
            try:
                TR = items.find_all('tr')
                for tr in TR:
                    tds = tr.find_all('td', {'class':'review-rating-header food_and_beverages'})
                    if tds:
                        star_fill_values = [span.text for span in tr.find('td',{'class' :
                        'review-rating-stars stars'}).find_all('span',{'class':'star fill'})]
                        total_star_fill = len(star_fill_values)
                        d['Food'] = total_star_fill
            except :
                d['food'] = None


            try:
                TR = items.find_all('tr')
                for tr in TR:
                    tds = tr.find_all('td', {'class':'review-rating-header inflight_entertainment'})
                    if tds:
                        star_fill_values = [span.text for span in tr.find('td',{'class' :
                        'review-rating-stars stars'}).find_all('span',{'class':'star fill'})]
                        total_star_fill = len(star_fill_values)
                        d['Inflight_entertainment'] = total_star_fill
            except :
                d['Inflight_entertainment'] = None
                
            try:
                TR = items.find_all('tr')
                for tr in TR:
                    tds = tr.find_all('td', {'class':'review-rating-header ground_service'})
                    if tds:
                        star_fill_values = [span.text for span in tr.find('td',{'class' :
                        'review-rating-stars stars'}).find_all('span',{'class':'star fill'})]
                        total_star_fill = len(star_fill_values)
                        d['Ground_service'] = total_star_fill
            except :
                d['Ground_service'] = None


            try:
                TR = items.find_all('tr')
                for tr in TR:
                    tds = tr.find_all('td', {'class':'review-rating-header wifi_and_connectivity'})
                    if tds:
                        star_fill_values = [span.text for span in tr.find('td',{'class' :
                        'review-rating-stars stars'}).find_all('span',{'class':'star fill'})]
                        total_star_fill = len(star_fill_values)
                        d['Wifi'] = total_star_fill
            except :
                d['Wifi'] = None
                
            try:
                TR = items.find_all('tr')
                for tr in TR:
                    tds = tr.find_all('td', {'class':'review-rating-header value_for_money'})
                    if tds:
                        star_fill_values = [span.text for span in tr.find('td',{'class' :
                        'review-rating-stars stars'}).find_all('span',{'class':'star fill'})]
                        total_star_fill = len(star_fill_values)
                        d['Value_for_money'] = total_star_fill
            except :
                d['Value_for_money'] = None

            #print(' ')   #To give space after each customer's review

            L.append(d)


```

```python
import numpy as np

import seaborn as sns

import matplotlib.pyplot as plt
%matplotlib inline
    
```

```python
df = pd.read_csv('britishReview.csv')
df
```

```python
df = df.drop(df.columns[:1],1)
df
```

```python
df.isnull().sum()
```

```python
df.dropna(inplace = True)
```

```python
df.isnull().sum()
```

```python
sns.heatmap(df.isnull(),yticklabels = False, cbar = False)
```

```python
sns.countplot(x = 'Recommended',data = df,palette = 'viridis')
plt.title('Count of Recommended vs. Non-Recommended')
plt.xlabel('Recommended')
plt.ylabel('Count')
plt.show()
```

```python
sns.countplot(x = 'Type_of_Traveller', hue = 'Wifi',data = df,palette = 'Blues')
plt.title('Count of Wifi Recommendation by Type of Traveller')
plt.xlabel('Type of Traveller')
plt.ylabel('Count')
plt.show()
```

```python
sns.boxplot(x ='Recommended', y= 'Seat_comfort',data = df, palette='pastel')
plt.title('Seat Comfort Ratings for Recommended vs Non-recommended Flights')
plt.xlabel('Recommended')
plt.ylabel('Seat Comfort Rating')

plt.show()
```

```python
sns.boxplot(x ='Recommended', y= 'Inflight_entertainment',data = df)
plt.title('Inflight_entertainment Ratings for Recommended vs Non-recommended Flights')
plt.xlabel('Recommended')
plt.ylabel('Inflight_entertainment Rating')

plt.show()
```

```python
sns.boxplot(x ='Recommended', y= 'Cabin_service',data = df)
plt.title('Cabin_service Ratings for Recommended vs Non-recommended Flights')
plt.xlabel('Recommended')
plt.ylabel('Cabin_service Rating')

plt.show()
```

```python
sns.pairplot(df, hue = 'Recommended', diag_kind = 'kde')
plt.suptitle('Pair Plot of Key Variables', y = 1.02)

plt.show()
```

```python
sns.countplot(x = 'Type_of_Traveller', hue = 'Recommended', data = df, palette = 'Set2')
plt.title('Count of Recommended vs Non-Recommended Flights by Type of Traveller')
plt.xlabel('Type of Traveller')
plt.ylabel('Count')

plt.show()
```
